With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = '/mnt/data/Air Quality data.csv'
data = pd.read_csv("/content/Air Quality data.csv")

# Clean column names: Strip leading/trailing spaces and remove special characters
data.columns = data.columns.str.strip().str.replace('.', '', regex=False)

# Check if key columns exist
required_columns = ["PM25 levels", "PM10 levels", "AQI bucket/category",
                    "CO (Carbon Monoxide) levels", "Date of the record",
                    "NOx levels", "SO2 (Sulfur Dioxide) levels", "O3 (Ozone) levels"]

for col in required_columns:
    if col not in data.columns:
        print(f"Column '{col}' is missing from the dataset. Skipping related plots.")

# Drop rows with missing values in critical columns
data = data.dropna(subset=["PM25 levels", "PM10 levels", "AQI bucket/category"])

# Set up visualization style
sns.set(style="whitegrid")
plt.figure(figsize=(14, 10))

# 1. Distribution of AQI categories
if "AQI bucket/category" in data.columns:
    plt.subplot(2, 2, 1)
    sns.countplot(x="AQI bucket/category", data=data, palette="coolwarm")
    plt.title("Distribution of AQI Categories")
    plt.xticks(rotation=45)

# 2. Scatter plot for PM2.5 vs PM10 levels
if "PM25 levels" in data.columns and "PM10 levels" in data.columns:
    plt.subplot(2, 2, 2)
    sns.scatterplot(x="PM25 levels", y="PM10 levels", hue="AQI bucket/category", data=data, palette="viridis")
    plt.title("PM2.5 vs PM10 Levels")

# 3. Line chart for gaseous pollutants
if "Date of the record" in data.columns and "NOx levels" in data.columns:
    plt.subplot(2, 2, 3)
    data.plot(x="Date of the record", y=["NOx levels", "SO2 (Sulfur Dioxide) levels", "O3 (Ozone) levels"],
              kind="line", ax=plt.gca())
    plt.title("Trends of NOx, SO2, and O3 Levels Over Time")
    plt.ylabel("Concentration (g/m)")
    plt.xlabel("Date")
    plt.legend(loc="upper left")

# 4. Boxplot for CO levels across AQI categories
if "CO (Carbon Monoxide) levels" in data.columns and "AQI bucket/category" in data.columns:
    plt.subplot(2, 2, 4)
    sns.boxplot(x="AQI bucket/category", y="CO (Carbon Monoxide) levels", data=data, palette="Set2")
    plt.title("CO Levels Across AQI Categories")
    plt.xticks(rotation=45)

# Display the plots
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean data
file_path = '/mnt/data/Air Quality data.csv'
data = pd.read_csv("/content/Air Quality data.csv")

# Clean column names
data.columns = data.columns.str.strip().str.replace('.', '', regex=False)

# Convert 'PM25 levels' to numeric, handling errors
data["PM25 levels"] = pd.to_numeric(data["PM25 levels"], errors='coerce')

# Drop rows with missing critical columns after conversion
data = data.dropna(subset=["Name of the city", "PM25 levels", "Air Quality Index (AQI) value"])

# Set up the visualization
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))

# 1. Bar Chart: Average PM2.5 Levels by City
city_pm25 = data.groupby("Name of the city")["PM25 levels"].mean().sort_values(ascending=False)
city_pm25.plot(kind="bar", color="skyblue")
plt.title("Average PM2.5 Levels by City", fontsize=16)
plt.xlabel("City", fontsize=12)
plt.ylabel("PM2.5 Levels (g/m)", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()

# 2. Scatter Plot: AQI vs PM2.5 for each City
plt.figure(figsize=(12, 6))
sns.scatterplot(
    x="PM25 levels",
    y="Air Quality Index (AQI) value",
    hue="Name of the city",
    data=data,
    palette="Set2",
    s=100
)
plt.title("PM2.5 Levels vs AQI by City", fontsize=16)
plt.xlabel("PM2.5 Levels (g/m)", fontsize=12)
plt.ylabel("Air Quality Index (AQI) Value", fontsize=12)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1), title="City")
plt.grid(axis="both", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
file_path = '/mnt/data/Air Quality data.csv'
data = pd.read_csv("/content/Air Quality data.csv")

# Clean column names
data.columns = data.columns.str.strip().str.replace('.', '', regex=False)

# Drop rows with missing critical columns
data = data.dropna(subset=["Name of the city", "AQI bucket/category"])

# Group data by AQI category
city_aqi_counts = data.groupby("AQI bucket/category")["Name of the city"].nunique()

# Pie chart
plt.figure(figsize=(8, 8))
colors = plt.cm.Paired(range(len(city_aqi_counts)))  # Generate colors dynamically
plt.pie(city_aqi_counts, labels=city_aqi_counts.index, autopct='%1.1f%%', colors=colors, startangle=140)

plt.title("City Distribution by AQI Categories", fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
file_path = '/mnt/data/Air Quality data.csv'
data = pd.read_csv("/content/Air Quality data.csv")

# Clean column names
data.columns = data.columns.str.strip().str.replace('.', '', regex=False)

# Check if 'Name of the city' column exists
if "Name of the city" not in data.columns:
    raise ValueError("The dataset does not contain a 'Name of the city' column.")

# Group data by city
city_counts = data["Name of the city"].value_counts()

# Pie chart
plt.figure(figsize=(8, 8))
colors = plt.cm.tab20c(range(len(city_counts)))  # Generate a dynamic color palette
plt.pie(
    city_counts,
    labels=city_counts.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=140
)

plt.title("Distribution of Air Quality Records by City", fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle
plt.tight_layout()
plt.show()
```
